```python
import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
import pandas as pd
```

```python
data = pd.read_csv("yanxi.txt",sep=";")
data.columns = ['username','rating','date','likes','comments']
data.head()
```

```python
data.sort_values(["rating"],inplace=True)
data.head()
```

```python
data.sort_values(["rating","likes"],inplace=True)
data.head()
```

https://www.jb51.net/article/86040.htm  多条件查询

```python
data5 = data.loc[data["rating"] == u"力荐"]
data5.sort_values("likes",inplace=True,ascending = False)
data5.head(10)
```

```python
data4 = data.loc[data["rating"] == u"推荐"]
data4.sort_values("likes",inplace=True,ascending = False)
data4.head(10)
```

```python
data3 = data.loc[data["rating"] == u"还行"]
data3.sort_values("likes",inplace=True,ascending = False)
data3.head(10)
```

```python
data2 = data.loc[data["rating"] == u"较差"]
data2.sort_values("likes",inplace=True,ascending = False)
data2.head(10)
```

```python
data1 = data.loc[data["rating"] == u"很差"]
data1.sort_values("likes",inplace=True,ascending = False)
data1.head(10)
```

测试有没有刷榜和恶意被黑的行为

```python
# value_counts()： https://zhuanlan.zhihu.com/p/29986906
# value_counts： https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html
data['comments'].value_counts().sort_index()
```

都是unique的，说明没有刷榜行为。
